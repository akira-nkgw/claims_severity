#### T-test
Since the distribution of 'loss' is not normally distributed, I decided to use Mann-Withney U-test.

```python
import scipy
```

```python
keys = []
tables = []
for variable in cat.columns[:72]:
    keys.append(variable)
    s, p = scipy.stats.mannwhitneyu(data['loss'].loc[data[variable] == 'A'], 
                                    data['loss'].loc[data[variable] == 'B'], 
                                    alternative='two-sided')
    tables.append(p) # append only p-value
df_t_test = pd.Series(tables, index = keys) 
df_t_test = df_t_test.where(df_t_test < .01).copy()
```

```python
tables
```

```python
df_t_test.keys()
```

The t-test with 1% criteria shows that all the variables cat1 to cat72 are all statistically significant. 

#### ANNOVA test

```python
import scipy.stats as stats
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
keys = []
tables = []
for variable in cat.columns[72:]:
    model = ols('loss ~ {}'.format(variable), data=data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    keys.append(variable)
    tables.append(anova_table['PR(>F)'][0]) # append only p-value

#df_anova = pd.concat(tables, keys=keys, axis=0)
df_anova = pd.Series(tables, index = keys) 
```

```python
df_anova
```

```python
df_anova < 0.01
```

Except for cat88, they are all statistically significant to loss at the 1% level significance
